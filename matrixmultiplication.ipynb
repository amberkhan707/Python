```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
plt.quiver(0, 0, i[0], i[1], angles='xy', scale_units='xy', scale=1, color='r', label="i")
plt.quiver(0, 0, j[0], j[1], angles='xy', scale_units='xy', scale=1, color='b', label="j")
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
t1 = np.array(([0,-1],
               [1,0]))
ri = t1 @ i
rj = t1 @ j
```

```python
plt.quiver(0, 0, ri[0], ri[1], angles='xy', scale_units='xy', scale=1, color='r', label="i")
plt.quiver(0, 0, rj[0], rj[1], angles='xy', scale_units='xy', scale=1, color='b', label="j")
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
t2 = np.array([[1,1],
               [0,1]])
si = t2 @ ri
sj = t2 @ rj
print(sj)
```

```python
plt.quiver(0, 0, si[0], si[1], angles='xy', scale_units='xy', scale=1, color='r', label="i")
plt.quiver(0, 0, sj[0], sj[1], angles='xy', scale_units='xy', scale=1, color='b', label="j")
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
